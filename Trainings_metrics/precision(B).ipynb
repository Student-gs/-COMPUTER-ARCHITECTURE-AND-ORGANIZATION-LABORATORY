```python
### 2. Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import os
```

## Metric of mAP50(B) over Epochs

```python
metric = 'metrics/precision(B)'
```

ADAM

```python
base_path = '/content/gdrive/MyDrive/YOLOTRAIN2'
adam_models = ['Adam_A1', 'Adam_A2', 'Adam_A3', 'Adam_A4']

plt.figure(figsize=(12, 6))

for model in adam_models:
    csv_path = os.path.join(base_path, model, 'train', 'results.csv')
    if os.path.exists(csv_path):
        df = pd.read_csv(csv_path)
        plt.plot(df[f"{metric}"], label=model)
    else:
        print(f"File not found: {csv_path}")

plt.title(f"Adam Optimizer Setups - {metric} per Epoch")
plt.xlabel("Epoch")
plt.ylabel(metric)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

SGD

```python
sgd_models = ['SGD_S1', 'SGD_S2', 'SGD_S3', 'SGD_S4']

plt.figure(figsize=(12, 6))

for model in sgd_models:
    csv_path = os.path.join(base_path, model, 'train', 'results.csv')
    if os.path.exists(csv_path):
        df = pd.read_csv(csv_path)
        plt.plot(df[f"{metric}"], label=model)
    else:
        print(f"File not found: {csv_path}")

plt.title("SGD Optimizer Setups - mAP@0.5 per Epoch")
plt.xlabel("Epoch")
plt.ylabel(metric)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
